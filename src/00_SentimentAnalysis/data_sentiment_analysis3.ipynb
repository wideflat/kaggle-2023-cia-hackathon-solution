```python
import pandas as pd
import numpy as np
from transformers import pipeline
```

```python
ID = 'tweet_id'
TARGET = 'engagement_count'
NROWS = None
DATA_DIR = "../input"
TRAIN_FILE = "{0}/expert_training.csv".format(DATA_DIR)
TEST_FILE = "{0}/expert_testing.csv".format(DATA_DIR)

# read data
train = pd.read_csv(TRAIN_FILE, nrows=NROWS)
test = pd.read_csv(TEST_FILE, nrows=NROWS)
```

```python
specific_model = pipeline(model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
text = list(train['full_text'])+list(test['full_text'])
new_text = [' '.join(item.split()[:128]) for item in text if item]
```

```python
nrows = 100

label = []
score = []
for i in range(round(len(new_text)/nrows+1)):
    print(f'\nfrom {nrows*i} to {nrows*(i+1)}...')
    try:
        sent_result = specific_model(new_text[nrows*i:nrows*(i+1)])
        print('   sucess! putting labels now...')
        for j in range(nrows):
            label.append(sent_result[j]['label'])
            score.append(sent_result[j]['score'])            
    except:
        print('   failed... putting labels now...')
        for j in range(nrows):
            label.append('NEU')
            score.append(0.5)
print('\n...Done')
```

```python
output = pd.DataFrame()

train_test = pd.concat([train['tweet_id'], test['tweet_id']], axis=0)
output['tweet_id'] = train_test
output = output.reset_index(drop=True)

output['sent_label'] = label[:len(train_test)]
output['sent_score'] = score[:len(train_test)]
```

```python
tmp = pd.get_dummies(output['sent_label'], prefix='sent_label', prefix_sep='_')
output = pd.concat([output, tmp], axis=1)
output = output.drop(['sent_label'], axis=1)
output.head()
```

```python
output.to_csv('sentiment_analysis_scores3.csv', index=False)
```
